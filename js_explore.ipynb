```python
#data1.host_response_rate.fillna(0,inplace=True)
data1.host_acceptance_rate.fillna(0,inplace=True)
data1.host_is_superhost.replace(['t','f'],[1,0], inplace=True)
data1.host_has_profile_pic.replace(['t','f'],[1,0],inplace=True)
data1.host_identity_verified.replace(['t','f'],[1,0],inplace=True)
data1.review_scores_value.fillna(0,inplace=True)
data1.require_guest_profile_picture.replace(['t','f'],[1,0],inplace=True)
data1.require_guest_phone_verification.replace(['t','f'],[1,0],inplace=True)
data1.instant_bookable.replace(['t','f'],[1,0],inplace=True)
data1.bathrooms.fillna(0,inplace=True)
data1.bedrooms.fillna(0,inplace=True)
data1.beds.fillna(0,inplace= True)
data1.reviews_per_month.fillna(0, inplace=True)


data1['price']=data1.price.apply(lambda x: x.replace('$',''))
data1['extra_people']=data1.extra_people.apply(lambda x: x.replace('$',''))
```
